# RFM-анализ клиентов магазина подарков

**Бизнес-кейс**: разработать дашборд  для проведения динамического RFM-анализа в BI-системе FineBI.

**Пользователи**: 
1. Маркетинговая команда — для разработки стратегии и настройки таргетированных коммуникаций.
2. Продуктовая команда и руководство — для мониторинга здоровья клиентской базы и принятия стратегических решений.

**Использованные данные:**

- Открытые данные о заказах клиентов онлайн-ритейла Online Retail - RFM&Cohort Analysis, Customer Seg (источник: Kaggle https://www.kaggle.com/datasets/bilalozgur/online-retail-rfm-and-cohort-analysis-customer-seg/data).
  Это набор данных, который содержит все заказы, произошедшие с 01.12.2010 по 09.12.2011 для зарегистрированного в Великобритании онлайн-ритейла. Компания в основном продает уникальные подарки на все случаи жизни. Многие клиенты компании являются оптовыми покупателями.
  
  
- Таблица "customer" содержит следующие поля:
1. `InvoiceDate` - день и время, когда была совершен каждый заказ;
2. `InvoiceNo` - номер заказа, если этот код начинается с буквы 'c', это указывает на отмену;
2. `StockCode` - код продукта;
3. `Description` - название продукта;
4. `Quantity` - количество каждого продукта в заказе;
5. `UnitPrice` - цена продукта за единицу в фунтах стерлингов;
6. `CustomerID` - уникальный номер клиента;
7. `Age` - возраст клиента;
8. `Gender` - пол клиента;
9. `Income` - доход;
10. `Country` - название страны, в которой проживает клиент;
  
**Вопросы исследования**:

RFM-анализ — метод анализа, позволяющий сегментировать клиентов по частоте и сумме покупок и выявлять тех, которые приносят больше денег.

Recency — давность (как давно клиенты совершали последнюю покупку). 

Frequency — частота (как часто клиенты покупают). Для расчета частоты нужно общее кол-во покупок разделить на кол-во лет.

Monetary — средний чек. 

RFM-анализ дает ответы на вопросы:

1. Кто лучшие клиенты? — выявление самых лояльных и прибыльных клиентов.
2. Кто в зоне риска? — обнаружение клиентов на грани оттока для своевременной реактивации.
3. Где скрытый потенциал? — определение сегментов, готовых к увеличению среднего чека. 
4. Кого нужно вернуть? — анализ потерянных клиентов.
5. Кого из клиентов необходимо удерживать? — приоритизация усилий по удержанию наиболее ценных групп.

Динамический RFM-анализ дает возможность настраивать сегментацию клиентов, меняя параметры на дашборде.

### Проверка наличия пропущенных значений, дубликатов, аномальных значений

Создадим функцию для предобработки данных `make_preprocess()`, с помощью которой найдем количество и процент пропущенных значений в столбце, выведем числовое описание столбца, для столбцов с типом данных `object` выведем уникальные значения и частоту значений в столбце.

```python
def make_preprocess(df):
    """Функция для предобработки данных находит количество и процент пропущенных значений
    в столбце, выводит числовое описание столбца,
    для столбцов с типом данных `object` выводит уникальные значения
    и частоту значений в столбце.

    Аргумент:
        датафрейм.
    """
    for column in df.columns:
        print(f"Cтолбец: {column}")
        print()
        print(
            f"Количество пропущенных значений (Null или NaN) {df[column].isna().sum()}"
        )
        print(f"Процент пропущенных значений {df[column].isna().mean().round(3):.1%}")
        print("*" * 90)
        print("Числовое описание столбца")
        print(df[column].describe())
        print("*" * 90)
        if df[column].dtype.name == "object":
            print("Уникальные значения в столбце")
            print(df[column].sort_values().unique())
            print("*" * 90)
            print("Частота в столбце")
            print(df[column].value_counts())
            print("*" * 90)
```

Применим функцию к датафрейму `sales`.

```python
make_preprocess(sales)
```

- Проверим наличие явных дубликатов и удалим их (если они есть) с помощью функции `get_duplicates()`.

```python
def get_duplicates(df):
    """Функция для проверки наличия явных дубликатов - выводит количество явных дубликатов,
    процент явных дубликатов или выводит сообщение, что явных дубликатов нет.

    Аргумент:
        датафрейм.
    """
    size = first_shape
    df_dupl = df.duplicated().sum()

    if df_dupl > 0:
        print(f"Количество явных дубликатов: {df_dupl}.")
        print(f"Процент явных дубликатов: {df_dupl/size:.1%} данных.")
    else:
        print("Явных дубликатов нет.")
```

- Посмотрим на среднее кол-во дней, прошедших с момента последней покупки.

```python
# максимальная дата в датасете
max_date = sales['invoice_date'].max()
max_date
```

```python
days_of_order_value = (
    sales
    .groupby('customer_id')
    .agg(days_from_order=('invoice_date', lambda x: (max_date - x.max()).days))
    .sort_values(by='days_from_order', ascending=True)
    .reset_index()
)

days_of_order_value.head()
```

```python
print("Среднее количество дней, прошедших с момента последней покупки:" ,
      round(days_of_order_value['days_from_order'].mean(),0))
print("Медианное количество дней, прошедших с момента последней покупки:" ,
      days_of_order_value['days_from_order'].median())
```

Т.к. среднее и медианное значения отличаются, в данных есть выбросы, значит будем использовать медианное значение для давности заказов на дашборде.

Для каждого клиента найдем стоимость и кол-во заказов.

```python
aov_value = (
    sales
    .groupby('customer_id')
    .agg(sales_agg=('total_sales','sum'), orders=('invoice_no','nunique'))
    .sort_values(by='sales_agg', ascending=False)
    .reset_index()
)

aov_value.head()
```

Найдем средний чек.

```python
aov_value['aov']=aov_value['sales_agg']/aov_value['orders']
aov_value.head()
```

```python
print("Среднее среднего чека: £", round(aov_value['aov'].mean(),0))
print("Медиана среднего чека: £", round(aov_value['aov'].median(),0))
```

Т.к. среднее и медианное значения отличаются, в данных есть выбросы, значит будем использовать медианное значение для среднего чека на дашборде.

### Создание параметров и расчетных полей

- Созданы параметры:

1. `param_F1` - для управления разбивкой на сегменты по Frequency (кол-во заказов больше этого попадает в группу "A")
2. `param_F2` - для управления разбивкой на сегменты по Frequency (кол-во заказов меньше или равно этому попадает в группу "С")
3. `param_R1` - для управления разбивкой на сегменты по Recency (кол-во дней с даты последней покупки меньше этого попадает в группу "A")
4. `param_R2` - для управления разбивкой на сегменты по Recency (кол-во дней с даты последней покупки больше или равно этому попадает в группу "С")
5. `param_M1` - для управления разбивкой на сегменты по Monetary (средний чек больше этого попадает в группу "A")
6. `param_M2` - для управления разбивкой на сегменты по Monetary (средний чек меньше или равный этому попадает в группу "С")
7. `param_group_period` - для управления разбивкой временной оси графика по кварталам и месяцам
8. `param_indicator_clients` - для управления выбором отображения кол-ва клиентов на графиках (абсолютное или в %)

### Сторителлинг

**RFM-анализ клиентов магазина подарков**

Проблема, которую мы решаем:
«У нас есть 3 920 клиентов и £7.285M выручки. Кто они? Кто приносит прибыль, а кто уходит? Как превратить случайную покупку в долгосрочную лояльность?»

Этот дашборд отвечает на вопросы бизнеса:
1. Кто лучшие клиенты? — выявление самых лояльных и прибыльных клиентов.
2. Кто в зоне риска? — обнаружение клиентов на грани оттока для своевременной реактивации.
3. Где скрытый потенциал? — определение сегментов, готовых к увеличению среднего чека. 
4. Кого нужно вернуть? — анализ потерянных клиентов.
5. Кого из клиентов необходимо удерживать? — приоритизация усилий по удержанию наиболее ценных групп.

#### Дизайн-решения 

Цветовая схема подходит для восприятия людьми с недостатком цветового зрения: синий, серый, красный, оранжевый.
Цвет используется для различия сегментов клиентов.

Фон дашборда не белоснежный (#fefefe), чтобы снизить контраст. Подложки визуальных элементов серые, используются для создания структуры дашборда.

Шрифт стандартный (Default), шрифт цифр на карточках (HarmonyOS). Цвет шрифта темно-серый и серый, чтобы не напрягать глаза от сильных контрастов. Используется иерархия размеров шрифта.

Принцип построения дашборда - от общего к частному.

Интерактивность:

При наведении на сегмент графика — подсветка + тултип с деталями.

Сегменты графиков и таблиц кликабельны (осуществляется фильтрация при нажатии на cегмент визуализации). Фильтрация из визуальных элементов первых двух вкладок распространяется на все вкладки. Фильтрация из визуальных элементов третьей вкладки распространяется только на третью вкладку. На третьей вкладе есть дополнительные фильтры для более точного поиска нужной информации.

Использованы параметры для динамического RFM-анализа (пользователь может управлять разбивкой клиентов на сегменты), для отображения кол-ва клиентов на графиках в абсолютных значениях и %, для отображения временной оси графика по кварталам и месяцам.

Есть кнопка "информация", при нажатии на которую открывается окно с пояснением для пользователя по работе с дашбордом.

#### Выводы и инсайты 

1. 19% клиентов (Champions + Loyal) генерируют 54% выручки. Принцип Парето в действии.

2. Наибольшее число клиентов (20% от всех) и выручки (40% от всей) было в ноябре-начале декабря 2011 года.

3. Средний чек Champions в два раза выше среднего чека всех клиентов.

4. Демографический портрет (3 920 клиентов, Великобритания):
    
        Женщины: 67% клиентов, 69% выручки, средний чек 275£
        Мужчины: 33% клиентов, 31% выручки, средний чек 286£ 

5. TOP-3 сегмента по выручке:

        Женщины 35-54 — «золотой сегмент»: 26% клиентов, 29% выручки. 
        Женщины 25-34: 22% клиентов, 20% выручки. 
        Мужчины 35-54: 13% клиентов, 16% выручки. 

6. TOP-3 сегмента клиентов-чемпионов:

        Женщины 25-34: 1,2% клиентов, средний чек 610£. Наибольшая лояльность, частота и средний чек.
        Женщины 35-54: 1,7% клиентов, средний чек 557£. 
        Женщины 55+: 0,6% клиентов, средний чек 552£. 

7. Потенциал

      Женщины 25-34 лет — наибольшая лояльность, частота и средний чек, нужно расширять базу таких клиентов.
      
      Мужчины 35-54 лет — давность чуть выше средней, чеки выше среднего.
